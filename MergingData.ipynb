```python
import numpy as np
import pandas as pd
```

```python
la_data = pd.read_csv("LA_data.csv")
c02_data = pd.read_csv("C02_Data.csv")
la_data = la_data.drop("Unnamed: 0",axis=1)
```

```python
print(c02_data)
```

```python
print(la_data)
```

```python
data = la_data.join(c02_data)
new_columns = ['Time', 'Temp (F)', 'Dew Point (F)', 'Humidity (%)', 'Wind', 'Wind Speed (mph)',
       'Wind Gust (mph)', 'Pressure (in)', 'Precip (in)', 'Condition', 'datetime_UTC', 'yyyy',
       'mm', 'dd', 'HH', 'first_valid_minute', 'last_valid_minute',
       'average_measurement_time', 'latitude', 'longitude', 'elevation_m',
       'intake_height_m', 'co2_ppm', 'co2_uncertainty', 'co2_SD_raw',
       'co2_n_raw', 'co2_SD_minutes', 'co2_n_minutes', 'co2_min', 'co2_max',
       'co2_flags']
data.columns = new_columns

data["Temp (F)"] = pd.to_numeric(data["Temp (F)"].str.replace(r' °F$', '', regex=True))
data["Dew Point (F)"] = pd.to_numeric(data["Dew Point (F)"].str.replace(r' °F$', '', regex=True))
data["Humidity (%)"] = pd.to_numeric(data["Humidity (%)"].str.replace(r'%$', '', regex=True))
data["Wind Speed (mph)"] = pd.to_numeric(data["Wind Speed (mph)"].str.replace(r' mph$', '', regex=True))
data["Wind Gust (mph)"] = pd.to_numeric(data["Wind Gust (mph)"].str.replace(r' mph$', '', regex=True))
data["Pressure (in)"] = pd.to_numeric(data["Pressure (in)"].str.replace(r' in$', '', regex=True))
data["Precip (in)"] = pd.to_numeric(data["Precip (in)"].str.replace(r' in$', '', regex=True))

data.drop(columns=['latitude', 'longitude', 'first_valid_minute', 'last_valid_minute', 'datetime_UTC', 'yyyy', 'mm', 'dd', 'HH'], inplace=True)

data.to_csv("c02_weather.csv")
```
